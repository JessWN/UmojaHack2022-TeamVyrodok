```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas_profiling import ProfileReport
sns.set()
```

```python
client_df = pd.read_csv('data/client_data.csv')
payment_df = pd.read_csv('data/payment_history.csv')
policy_df = pd.read_csv('data/policy_data.csv')
train_df = pd.read_csv('data/train.csv')
smp_sub_df = pd.read_csv('data/sample_sub.csv')
```

```python
def create_pd_profile_report(df, df_title = ''):
    profile = ProfileReport(df, title = '{} Pandas Profile Report'.format(df_title), explorative=True)
    profile.to_file('{}_ProfileReport.html'.format(df_title))
```

```python
create_pd_profile_report(client_df, 'Client dataset')
create_pd_profile_report(policy_df, 'Policy dataset')
create_pd_profile_report(pay_hist_df, 'Payment History dataset')
create_pd_profile_report(sample_df, 'Sample dataset')
create_pd_profile_report(train_df, 'Train dataset')
# client_profile
```
